## Como é a distribuição do peso dos produtos no conjunto de dados?

 Histograma com densidade para mostrar a distribuição do peso dos produtos.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df_novo_olist['product_weight_g'], bins=30, kde=True, color='skyblue')
plt.title('Distribuição do Peso dos Produtos')
plt.xlabel('Peso (g)')
plt.ylabel('Densidade')
plt.show()
```

## Como os preços dos produtos estão distribuídos no conjunto de dados?

Histograma com densidade para mostrar a distribuição dos preços dos produtos.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_novo_olist['price'], bins=30, kde=True, color='salmon')
plt.title('Distribuição dos Preços dos Produtos')
plt.xlabel('Preço')
plt.ylabel('Densidade')
plt.show()
```

## Como a distribuição do peso dos produtos afeta o valor do frete?

Gráfico de dispersão para explorar a relação entre o peso dos produtos e os valores de frete.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='price', y='freight_value', data=df_novo_olist, color='orange')
plt.title('Relação entre Preços dos Produtos e Valores do Frete')
plt.xlabel('Preço do Produto')
plt.ylabel('Valor do Frete')
plt.show()

```

## Como a localização geográfica dos vendedores influencia os valores do frete?

Um gráfico de barras apresentará a média dos valores do frete para cada estado dos vendedores.

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='seller_state', y='freight_value', data=df_novo_olist, palette='viridis', ci=None)
plt.title('Média dos Valores do Frete por Estado do Vendedor')
plt.xlabel('Estado do Vendedor')
plt.ylabel('Média do Valor do Frete')
plt.show()

```

Analisando os resultados obtidos, é possível extrair insights sobre a dinâmica dos valores de frete, distribuição de peso dos produtos e precificação no conjunto de dados. A influência da localização geográfica dos vendedores nos valores do frete tornou-se evidente, destacando variações significativas entre os estados. O gráfico de barras da média dos valores do frete por estado proporcionou uma visão clara dessa relação.

Ao explorar a distribuição do peso dos produtos, observamos que a maioria dos produtos tem pesos concentrados em determinadas faixas, proporcionando uma compreensão mais precisa de como o peso impacta os custos de frete. A análise da relação entre os preços dos produtos e os valores do frete revelou padrões interessantes, destacando áreas onde os custos de frete podem estar mais sensíveis aos preços dos produtos.

A distribuição dos preços dos produtos, como evidenciado pelo histograma, apresentou uma variedade de faixas de preços, indicando a presença de produtos em diferentes segmentos de mercado. Essa diversidade nos preços sugere a possibilidade de estratégias de precificação distintas dentro do conjunto de dados.

Logo, essas análises oferecem uma visão abrangente das relações complexas entre localização geográfica, peso dos produtos e precificação, fornecendo insights valiosos para a tomada de decisões estratégicas no contexto do comércio eletrônico.
